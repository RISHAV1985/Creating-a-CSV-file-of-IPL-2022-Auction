```python
# Importing required libraries 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
# loading dataset from github
url="https://raw.githubusercontent.com/RISHAV1985/Creating-a-CSV-file-of-IPL-2022-Auction/main/ipl_2022_dataset_rishav.csv"
data=pd.read_csv(url)
```

```python
# Details of dataset
data.shape
data.columns
```

```python
# representation of dataset
data.sample()
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# List of Players who went unsold
data[data['Cost IN $ (000)'].isnull()]
# Making the cost of unsold players to zero
data['COST IN ₹ (CR.)']=data['COST IN ₹ (CR.)'].fillna(0)
data['Cost IN $ (000)']=data['Cost IN $ (000)'].fillna(0)
```

```python
# Players who eiher went unsold or participated first time in ipl auction
data[data['2021 Squad'].isnull()]
```

```python
# Making null value to not participated term
data['2021 Squad']=data['2021 Squad'].fillna('Not Participated in IPL 2021')
```

```python
# Replacing all null values 
data.isnull().sum()
```

```python
# adding and adjusting new columns for detailed analysis
teams = data[data['COST IN ₹ (CR.)']>0]['Team'].unique()
teams
```

```python
data['status']=data['Team'].replace(teams,'sold')
data['Base Price'].unique()
```

```python
data['retention']=data['Base Price']
```

```python
data['retention'].replace(['2 Cr', '40 Lakh', '20 Lakh', '1 Cr', '75 Lakh',
       '50 Lakh', '30 Lakh','1.5 Cr'],'In Auction', inplace = True)
```

```python
# Dealing with base price column
data['Base Price'].replace('Draft Pick',0, inplace = True)
data['base_price_unit'] = data['Base Price'].apply(lambda x: str(x).split(' ')[-1])
data['base_price'] = data['Base Price'].apply(lambda x: str(x).split(' ')[0])
data['base_price'].replace('Retained',0,inplace=True)
data['base_price_unit'].unique()
```

```python
data['base_price_unit'] = data['base_price_unit'].replace({'Cr':100,'Lakh':1,'Retained':0})
data['base_price'] = data['base_price'].astype(float)
data['base_price_unit'] = data['base_price_unit'].astype(int)
data['base_price'] = data['base_price']*data['base_price_unit']
data.head()
```

```python
data.drop(['Base Price','base_price_unit'], axis =1, inplace = True)
```

```python
data
```

```python
data['COST IN ₹ (CR.)'] = data['COST IN ₹ (CR.)']*100
data = data.rename(columns={'TYPE':'Type','COST IN ₹ (CR.)':'Sold_for_lakh','Cost IN $ (000)':'Cost_in_dollars','2021 Squad':'Prev_team','Team':'Curr_team'})
```

```python
data.head()
```

```python
# Checking for duplicate Players
data[data['Player'].duplicated(keep=False)]
```

The above list looks like of duplicate players but all players are different.

Now our dataset is ready for analysis.

```python
# Total number of based on their roles(Batsman, Bowler, Allrounder and Wicketkeeper)
role=data['Type'].value_counts()
role.reset_index()
```

```python
# Pictorial Representation of above data
plt.pie(role.values, labels=role.index,labeldistance=1.2,autopct='%1.2f%%')
plt.title('Role of Players Participated', fontsize = 15)
plt.plot()
```

```python
plt.figure(figsize=(4,4))
fig = sns.countplot(data['status'],palette=['Green','Red'])
plt.xlabel('Sold or Unsold')
plt.ylabel('Number of Players')
plt.title('Sold vs Unsold', fontsize=15)
plt.plot()

for p in fig.patches:
    fig.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width()/2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')
```

```python
# Representation of bought players by individual teams 
plt.figure(figsize=(20,10))
fig = sns.countplot(data[data['Curr_team']!='Unsold']['Curr_team'])
plt.xlabel('Name of Team')
plt.ylabel('Number of Players')
plt.title('Players Brought by each Team', fontsize=15)
plt.xticks(rotation=90)
plt.plot()

for p in fig.patches:
    fig.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width()/2., p.get_height()), ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')
```
